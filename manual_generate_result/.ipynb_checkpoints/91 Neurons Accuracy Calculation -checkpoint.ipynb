# 输出不同列表中的唯一对例子

```python
from itertools import combinations

# 定义你的列表
list1 = [14,15,17]
list2 = [9,10,20]
list3 = [6,7,18]
list4 = [2,7,18]
list5 = [1,4,11]

# 将所有列表放到一个大列表里
lists = [list1, list2, list3, list4, list5]

# 对于大列表中的每一个列表，使用itertools.combinations找到所有的两两组合
for i, lst in enumerate(lists):
    unique_pairs = list(combinations(lst, 2))
    print(f'在 list{i+1} 中的唯一组合是: {unique_pairs}')
```

```python
from itertools import combinations

# 定义你的列表
list1 = [14,15,17]
list2 = [9,10,20]
list3 = [6,7,18]
list4 = [2,7,18]
list5 = [1,4,11]

# 将所有列表放到一个大列表里
lists = [list1, list2, list3, list4, list5]

# 使用一个set存储所有的唯一对
unique_pairs_set = set()

# 对于大列表中的每一个列表，使用itertools.combinations找到所有的两两组合
for lst in lists:
    unique_pairs_set.update(combinations(lst, 2))

# 为了方便查看，可以转换成列表并打印
unique_pairs_list = list(unique_pairs_set)
print(f'所有列表中的唯一对（没有重复）是: {unique_pairs_list}')
```

# 输出不同列表中的唯一对应用

```python
from itertools import combinations

# 定义你的列表
list1 = [14,15,17,28,29,30,31,33,47,48,49,51,52,76,80]
list2 = [9,10,20,21,22,39,40,41,63,64,8,38,65,66]
list3 = [6,7,16,17,32,33,34,53,54,55,82,84,56]
list4 = [2,7,18,19,35,36,37,59,60,61,62,88,90]
list5 = [1,4,11,12,13,23,24,25,26,44,45,46,69,70]

# 将所有列表放到一个大列表里
lists = [list1, list2, list3, list4, list5]

# 使用一个set存储所有的唯一对
unique_pairs_set = set()

# 对于大列表中的每一个列表，使用itertools.combinations找到所有的两两组合
for lst in lists:
    pairs = combinations(lst, 2)
    sorted_pairs = [tuple(sorted(pair)) for pair in pairs] 
    unique_pairs_set.update(sorted_pairs)

# 为了方便查看，可以转换成列表并打印
grand_truth = list(unique_pairs_set)
print(f'所有列表中的唯一对（没有重复）是: {grand_truth}')
```

# 比较重复元素的数量和列表的总长度来计算重复度的比率

```python
# 定义你的列表
list1 = [(9, 10), (6, 18), (2, 7), (14, 17), (7, 18)]
list2 = [(9, 8), (6, 18), (2, 7), (4, 17), (7, 18)]

# 将列表转换为集合
set1 = set(list1)
set2 = set(list2)

# 计算两个集合的交集
common = set1 & set2

# 计算重复度的比率
duplication_ratio = len(common) / min(len(set1), len(set2))

# 输出比率
print(f'重复度的比率是: {duplication_ratio}')
```
