```python
import matplotlib.pyplot as plt
import numpy as np

# 假设有三组数据集
data1 = np.random.normal(loc=10, scale=2, size=50)
data2 = np.random.normal(loc=15, scale=3, size=40)
data3 = np.random.normal(loc=8, scale=1, size=60)

# 计算每组数据的均值
mean1 = np.mean(data1)
mean2 = np.mean(data2)
mean3 = np.mean(data3)

# 绘制散点图
plt.scatter(1, mean1, color='b', label='Dataset 1')
plt.scatter(2, mean2, color='g', label='Dataset 2')
plt.scatter(3, mean3, color='r', label='Dataset 3')

# 设置图表标题和坐标轴标签
plt.title('Mean Scatter Plot')
plt.xlabel('Dataset')
plt.ylabel('Mean Value')

# 显示图例
plt.legend()

# 设置x轴刻度
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 假设每组数据有10个数据点
data1 = np.random.normal(loc=10, scale=2, size=10)
data2 = np.random.normal(loc=15, scale=3, size=10)
data3 = np.random.normal(loc=8, scale=1, size=10)

# 计算每组数据的均值
mean1 = np.mean(data1)
mean2 = np.mean(data2)
mean3 = np.mean(data3)

# 绘制散点图
plt.scatter(np.ones(10), data1, color='r', label='Dataset 1')
plt.scatter(2 * np.ones(10), data2, color='r', label='Dataset 2')
plt.scatter(3 * np.ones(10), data3, color='r', label='Dataset 3')

# 绘制均值的红色散点
plt.scatter(1, mean1, color='green', marker='o', s=100, label='Mean 1')
plt.scatter(2, mean2, color='yellow', marker='o', s=100, label='Mean 2')
plt.scatter(3, mean3, color='blue', marker='o', s=100, label='Mean 3')

# 设置图表标题和坐标轴标签
plt.title('Mean Scatter Plot')
plt.xlabel('Dataset')
plt.ylabel('Value')

# 显示图例
plt.legend()

# 设置x轴刻度
plt.xticks([1, 2, 3], ['Dataset 1', 'Dataset 2', 'Dataset 3'])

# 显示图表
plt.show()
```
